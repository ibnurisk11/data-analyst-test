### Import the required library

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns

# Load the mtcars dataset
df = sns.load_dataset(“mpg”) # Alternative to mtcars in Python
df = df.dropna() # Remove rows with NaN values
```

### Checking Columns

```python
print(df.columns)  # Cek apakah nama kolom sudah sesuai
```

### Selecting Variables

```python
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']] # predictor variables
y = df['mpg'] # target variable
```

### Adding Constants & Building a Model

```python
X = sm.add_constant(X)  # Add a constant for the regression model
model = sm.OLS(y, X).fit()
print(model.summary())  # Display model summary
```

### Results Analysis

```python
from sklearn.linear_model import LinearRegression

# Build regression model with all data
model = LinearRegression()
model.fit(X, y)

# Display regression coefficients and intercept
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
```

### Conclusion on the Model

The linear regression model was built to predict the `mpg` (miles per gallon) of vehicles based on features such as `cylinders`, `displacement`, `horsepower`, `weight`, and `acceleration`. Below are the key points to consider:

1. **Model Fit**: 
    - The model summary from the `statsmodels` library provides statistical metrics such as R-squared, p-values, and coefficients. These metrics help evaluate how well the model fits the data.
    - A high R-squared value indicates that the model explains a significant portion of the variance in the target variable (`mpg`).

2. **Feature Importance**:
    - The coefficients of the predictors indicate their impact on `mpg`. Positive coefficients suggest a direct relationship, while negative coefficients suggest an inverse relationship.
    - For example, features like `weight` and `horsepower` might have negative coefficients, indicating that heavier and more powerful cars tend to have lower `mpg`.

3. **Model Performance**:
    - The model was trained and tested using a train-test split. The performance on the test set (e.g., using metrics like Mean Squared Error or R-squared) indicates how well the model generalizes to unseen data.

4. **Limitations**:
    - The model assumes a linear relationship between predictors and the target variable, which might not always hold true.
    - Multicollinearity among predictors (e.g., `weight` and `horsepower`) could affect the reliability of the coefficients.

5. **Next Steps**:
    - Evaluate residuals to check for patterns that might indicate non-linearity or heteroscedasticity.
    - Consider feature engineering or using more complex models (e.g., polynomial regression or tree-based models) if the linear model does not perform well.

Overall, the model provides a baseline understanding of the relationship between vehicle features and fuel efficiency, but further refinement and validation are necessary for better predictions.
